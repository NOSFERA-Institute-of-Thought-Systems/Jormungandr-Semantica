```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configure plotting style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
# Load the data generated by the experiment script
data_path = Path("../outputs/theorem_4_3_validation_data.csv")
df = pd.read_csv(data_path)

print("Data loaded successfully:")
df.head()
```

```python
fig, ax1 = plt.subplots()

# Plot Spectral Gap (lambda_2) on the left y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Clique Size (m)', fontsize=14)
ax1.set_ylabel('Spectral Gap (λ₂)', color=color1, fontsize=14)
ax1.plot(df['m'], df['lambda_2'], color=color1, marker='o', linestyle='-', label='Spectral Gap (Empirical)')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(False, axis='y')

# Create a second y-axis for the curvature
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Bridge Curvature (κ)', color=color2, fontsize=14)
ax2.plot(df['m'], df['kappa_bridge'], color=color2, marker='s', linestyle='--', label='Bridge Curvature (Empirical)')
ax2.tick_params(axis='y', labelcolor=color2)
ax2.grid(False)

# Add titles and legends
fig.suptitle('Empirical Validation of Theorem 4.3\nInverse Relationship between Bridge Curvature and Spectral Gap', fontsize=16)
fig.tight_layout(rect=[0, 0, 1, 0.96])

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='center right')

# Save the figure for the paper
plt.savefig('../paper/figures/theorem_4_3_validation.png', dpi=300, bbox_inches='tight')

plt.show()
```
